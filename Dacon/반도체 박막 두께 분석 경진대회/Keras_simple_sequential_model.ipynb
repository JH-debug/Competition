```python
import pandas as pd
import numpy as np
from scipy.stats import norm
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import os

train = pd.read_csv(work_dir + "train.csv")
test = pd.read_csv(work_dir + "test.csv")

print(train.shape)
print(test.shape)
```

```python
fea_train = train.iloc[:,4:]  #Train 데이터의 feature
tar_train = train.iloc[:,0:4] #Train 데이터의 target
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout, BatchNormalization
from keras.callbacks import ModelCheckpoint

MODEL_SAVE_FOLDER_PATH = work_dir 
if not os.path.exists(MODEL_SAVE_FOLDER_PATH): 
    os.mkdir(MODEL_SAVE_FOLDER_PATH)
    
model_path = MODEL_SAVE_FOLDER_PATH + '{epoch:02d}_{acc:.4f}.hdf5' 
cb_checkpoint = ModelCheckpoint(filepath=model_path, monitor='acc', verbose=1, save_best_only=True)

model = Sequential()
model.add(Dense(units=fea_train.shape[1], activation='relu', input_dim=fea_train.shape[1], kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))
model.add(Dense(units=fea_train.shape[1], activation='relu', kernel_initializer='he_normal'))

model.add(Dense(units=4, activation='linear', kernel_initializer='he_normal'))
```

```python
model.compile(loss='mae', optimizer='adam', metrics=['acc']) #mean_absolute_error>>mse
model.fit(fea_train, tar_train, epochs=1500, batch_size=1000,callbacks=[cb_checkpoint]) # validation_split = 0.05)
```

```python
# loss 

loss = model.evaluate(fea_train , tar_train, batch_size = 32)
print('loss: ', str(loss))
```

```python
fea_test = test.iloc[:,1:]
```

```python
pred_test=model.predict(fea_test)
```

```python
sample_sub = pd.read_csv(work_dir + 'sample_submission.csv', index_col=0)
submission = sample_sub+pred_test
submission.to_csv(work_dir +'submission_9.csv')
```
